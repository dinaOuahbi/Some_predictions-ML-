##FILTRING DATA

```python
#isoler les type grass dans un nouveau df nomer 'grass'
grass = df.loc[df['Type 1']=='Grass']
#avec 
grassP = df.loc[(df['Type 1']=='Grass') & (df['Type 2']=='Poison')]
grassP
grassP70 = df.loc[(df['Type 1']=='Grass') & (df['Type 2']=='Poison') & (df['HP']>70)]
grassP70.shape
```

```python
grassP70.to_csv('grassP70.csv', sep='')
```
