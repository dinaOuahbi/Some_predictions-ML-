```python
import matplotlib.pyplot as plt
df['class'].value_counts().plot.bar()
plt.title('effectif number of each class')
plt.xlabel('class')
plt.ylabel('effectif')
```

```python
#en histograme on a la lomgeur sepale en fn du nb d'effectif avec une affinity de 50
col = 'sepal length (cm)'
df[col].hist(bins=15)
plt.title(col)
plt.xlabel(col)
plt.ylabel('Effectif')
plt.show()
```

```python
col = 'petal length (cm)'
df[col].hist(bins=50)
plt.title(col)
plt.xlabel(col)
plt.ylabel('Effectif')
plt.show()
```
