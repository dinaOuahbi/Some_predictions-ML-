INTERPOLATE NAN DATA

```python
import numpy as np 
import matplotlib.pyplot as plt
#11 points entre 0 et 10 
x = np.linspace(0, 10, 11)
y = x**2
```

```python
plt.scatter(x, y)
```

```python
#interpoler les donner new_x sur x
from scipy.interpolate import interp1d
f = interp1d(x, y, kind='linear')
new_x = np.linspace(0, 10, 10)
result = f(new_x)
```

```python
#afficher les resultats d'interpolation 
plt.scatter(x, y, label='x')
plt.scatter(new_x, result, c='r', label='new_x')
plt.legend()
```

OPTIMIZE/MINIMIZE

```python
X = np.linspace(0, 2, 100)
y = 1/3 * X**3 - 3/5 * X**2 + 2  + np.random.randn(100)/20
plt.scatter(X, y)
X.shape[0]
```

```python
#crrer un model de degre 3 que l'on va optimiser
def f(x,a,b,c,d):
    return a*x**3 + b*x**2 + c*x + d
```

```python
from scipy import optimize
params, param_cov = optimize.curve_fit(f, X, y)
#renvoie deux tab : 
    #un tableau avec les parametres du model (a,b,c,d)
    #un tableau avec les parametres de convergence 
#on a stoquer les deux tableau dans des variables
```
