```python
df.drop(['malic_acid', 'flavanoids', 'class', 'od280/od315_of_diluted_wines', 'total_phenols', 'proanthocyanins', 'ash', 'hue', 'nonflavanoid_phenols'], axis='columns', inplace=True)
df.head()
```
