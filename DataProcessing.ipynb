```python
import numpy as np
from sklearn import datasets
#telecharger les donnees iris
iris_X, iris_y = datasets.load_iris(return_X_y=True)
#envoie les target sans repetition
np.unique(iris_y)
```

```python
import numpy as np 
#initier le generateur de nb aleatoire fixe
np.random.seed(0)
#lancer le generteur pour 5 nombre
np.random.rand(5)
```

```python
#separer les caracteristiaues des cibles
import matplotlib.pyplot as plt
X = iris.data[:,:2]
y = iris.target
#representer les donnees
plt.figure(2, figsize=(8, 6))
plt.scatter(X[:,0], X[:,1], c=y, cmap=plt.cm.Set1, edgecolor='k', s=40)
plt.xlabel('sepal lenght')
plt.ylabel('sepal with')
plt.show()
```

```python
#figure 3d
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(1, figsize=(8, 6))
ax = Axes3D(fig, elev=-150, azim=110)
ax.scatter(X[:,0], X[:,1],iris.data[:,2], c=y, cmap=plt.cm.Set1, edgecolor='k', s=40)
ax.set_xlabel(iris.feature_names[0])
ax.set_ylabel(iris.feature_names[1])
ax.set_zlabel(iris.feature_names[2])
```

split iris data in train and test data

```python
#melangeur notre data de iris
indices = np.random.permutation(len(iris_X))
```

```python
#je malange et je prend les 10 premieres lignes
X_train = iris_X[indices[:-10]]
y_train = iris_y[indices[:-10]]
#je malange et je prend les 10 dernieres lignes
X_test = iris_X[indices[-10:]]
y_test = iris_y[indices[-10:]]
```

```python
#je cree le model
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train, y_train)
```

```python
knn.predict(X_test)
```

```python
y_test
```

```python
knn.score(X_test, y_test)
```

```python
#PREDICT DESEASE PROFGRESSION FROM DIABETES DATASETS
dt= datasets.load_diabetes()
```

```python
dt.feature_names
dt.target
#442 patients et 10 caracteristiaues physiologiques
dt.data.shape
```

```python
dt_X_train = dt.data[:-20]
dt_X_test = dt.data[-20:]
dt_y_train = dt.target[:-20]
dt_y_test = dt.target[-20:]
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(dt_X_train, dt_y_train)
```

```python
dt_y_pred = reg.predict(dt_X_test)
```

```python
#calculer l erreur
np.mean = ((dt_y_pred - dt_y_test)**2)
#np.mean
```

```python
#caluler le score
sc = reg.score(dt_X_test, dt_y_test)
sc
```
