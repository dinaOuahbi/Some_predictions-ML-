```python
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt
```

##simple regression with one feature 

```python
plt.figure()
X_R1, y_R1 = make_regression(n_samples = 100, n_features=1, n_informative=1, bias=150, noise=20, random_state=0)
```

```python
plt.scatter(X_R1, y_R1, marker='o', s=50)
plt.xlabel('X_R1')
plt.ylabel('y_R1')
plt.title('classification with one feature')
```

##simple binary clssification with two features


```python
from sklearn.datasets import make_classification
```

```python
plt.figure()
X_C2, y_C2 = make_classification(n_samples = 100, n_features=2,n_redundant=0, n_informative=2, n_clusters_per_class=1,flip_y=0.1, class_sep=0.5, random_state=0)
```

```python
plt.scatter(X_C2[:,0], X_C2[:,1],c=y_C2, marker='o', s=50)
plt.xlabel('first feature')
plt.ylabel('seconde feature')
plt.title('linear classification with two features')
```

##sample binary classif with non-linearly separable classes

```python
from  sklearn.datasets import make_blobs
```

```python
#geenrer 100 echantillon dans 8 grappes dfferentes et deux features
X_D2, y_D2 = make_blobs(n_samples = 100, n_features=2, centers=8, cluster_std=1, random_state=4)
```

```python
#modulo va convertir le nombre de y assigner par 'blobs' (entre 1 et 8)en un nombre binaire
#les point d'index paire = class 0 
#les point d'index impaire = class 1
y_D2%=2
plt.figure()
plt.scatter(X_D2[:,0], X_D2[:,1], c=y_D2, marker='o', s=50)  
plt.xlabel('first feature')
plt.ylabel('seconde feature')
plt.title('non-linear classification with two features')
```
